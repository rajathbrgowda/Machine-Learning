### Visualization of dataset

```python
#using seaborn library
#graphical analysis

sns.pairplot(iris, hue='species', height=4, aspect=1)  # dataset, groupBy, graphHeight(inches), aspectRatio
```

```python
#histogram
iris.hist(edgecolor='#FBF9F9',color='#CC9A8F',linewidth=1.2, figsize=(12,8));
plt.show();
```

```python
#using violin plot to show statisticala analysis of data
plt.figure(figsize=(16,12));
plt.subplot(2,2,1)
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='species', y='sepal_width', data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='species', y='petal_length', data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='species', y='petal_width', data=iris)
```

```python
#using boxplot
iris.boxplot(by='species', figsize=(16,12))
```

```python
#usnig scatter plot
pd.plotting.scatter_matrix(iris, figsize=(16,12))
plt.show()
```

```python
sns.pairplot(iris,hue='species', diag_kind='kde')
```
